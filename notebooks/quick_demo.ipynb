# CodeLens QA Demo

```python
import sys
sys.path.append('../src')
from codelens.ast_indexer import index_repo
from codelens.query_pipeline import QueryPipeline
```

```python
# Index the sample repo
repo_path = '../examples/sample_repo'
units = index_repo(repo_path)
print(f"Indexed {len(units)} units")
```

```python
# Initialize Pipeline
pipeline = QueryPipeline(units)
```

```python
# Ask a question
q = "How is data processed?"
result = pipeline.run(q)
import json
print(json.dumps(result, indent=2))
```
